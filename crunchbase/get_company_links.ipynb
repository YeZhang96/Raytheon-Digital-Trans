```python
from bs4 import BeautifulSoup
import os
import re
from IPython import embed

path = "./htmls/"
html_files = [f for f in os.listdir(path) if f.endswith(".html")]
print(len(html_files))
```

```python
links = []
for html_file in html_files:
    class_name = re.compile("cb-link component--field-formatter.*")
    f = open(path+html_file)
    soup = BeautifulSoup(f, "html.parser")
    fields = soup.find_all("a", {"class" : class_name})
    links += [i["href"] for i in fields if i["href"].startswith('/organization')]
    f.close()
links = set(links)
for l in links:
    print(l)
```
